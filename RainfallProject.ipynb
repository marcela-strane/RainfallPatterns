# Welcome to the rainfall Project's Code.
Step 1: Add your CSV file and press play to do initial setup

```python
import pandas as pd
from google.colab import files
filepath= '6HourData.csv'#Change this filepath to use a different dataset
data = pd.read_csv(filepath)
# reading the file and importing set up of the rainfall data
# Rain gauge #43201 was deleted from this file to reduce the data size since it was 0 across all years.

```

# Step 2: Calculate all rainfall events of an hour or more
Follow the comments to change the interval. Output will be in the same format as initial csv file and will contain all rainfall events within the the parameters.

```python
sums = data.sum(axis=1)
i=0
while i < len(data): #while used to be able to change main index
  Events = []#Create New Event
  if(sums[i]):#if event found continue until full rainfall event is collected
    Events.append(data.iloc[i])
    for j in range(i+1,len(data)):
      i=j
      if(sums[j]):
        Events.append(data.iloc[j])
      else:
        if len(Events):
          events = pd.DataFrame(Events)
          name = Events[0][0].replace("/", "-") + '.csv'#Download newfound rainfall event
          events.to_csv(name) 
          files.download(name)
        i=j
        break

  i+=1#increment index
  #print(Events)
```

```python
#Generate CSVs and download to machine
events = pd.DataFrame(Events)
events.to_csv('Events Rainfall Data2017-2019.csv') 
files.download('Events Rainfall Data2017-2019.csv')
```
